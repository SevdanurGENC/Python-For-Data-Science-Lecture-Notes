```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Iris.csv")
```

```python
sample_data = data.sample(20)
```

```python
for _ in range(10):
    sample_data.iloc[np.random.randint(1,20), np.random.randint(1,5)] = np.nan
```

```python
all_data = pd.concat([data, sample_data], axis=0)
```

```python
selected_columns = data.columns[1:-1]
```

```python
all_data.sort_values(by="Species", inplace=True)
```

```python
all_data.reset_index(drop=True, inplace=True)
all_data.drop("Id", axis=1, inplace=True)
```
